```python
df1=pd.read_csv('/content/Phising_Training_Dataset.csv',index_col='key')
df1.head()
```

```python
df2=pd.read_csv('/content/Phising_Testing_Dataset.csv',index_col='key')
df2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X = df1.iloc[:, :-1]
y = df1.Result
X_test=df2.copy()
X_train, X_valid, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
model = DecisionTreeClassifier()

model.fit(X_train, y_train)
```

```python
ypred = model.predict(X_test)
```

```python
df_output=pd.DataFrame()
df_output['Key']=X_test.index
df_output['Result']= ypred
df_output[['Key','Result']].to_csv('output1.csv',index=False)
```
